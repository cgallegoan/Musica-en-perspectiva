```python
import pandas as pd
df = pd.read_csv('outputpop.csv', sep=',', encoding = 'ISO-8859-1')
df.head()
```

```python
import re
```

```python
c = 0
for valor in df['artists']:
    valor = re.split(r',',valor)
    valor = valor[0]
    valor = re.findall(r'\s*\w+\s*',valor)
    res = ''
    for v in valor:
        res += v
    valor = res
    df['artists'][c] = valor
    c += 1
    if c== 100 or c == 1000 or c == 100000 or c== 250000 or c == 500000 or c == 750000 or c== 1000000 or c == 1250000:
        print(c)
df.head()
```

```python
df.to_csv('datospopart1.csv', sep = ',')
```
